```python
# Tutaj przygotowuję sobie słowniki składające się z klucz(ilość wystąpień) : wartość[słowo, następnik1, następnik2]
trigrams_dict = {}
with open('./../ruznezecy/poleval_3grams.txt', 'r', encoding='utf8') as trigrams:
    for line in trigrams:
        current_line = line.split()
        trigrams_dict[current_line[0]] = current_line[1:4]
sorted_trigrams_dict = {k: v for k, v in sorted(trigrams_dict.items(), key=lambda item: item[1])}
```

```python
trigrams_results_dict = {}
for key in sorted_trigrams_dict:
    arr = sorted_trigrams_dict[key]
    if arr[0] in trigrams_results_dict.keys():
        if len(arr) == 3:
            trigrams_results_dict[arr[0]].append([arr[1], arr[2]])
        else:
            trigrams_results_dict[arr[0]].append(arr[1])
    else:
        if len(arr) == 3:
            trigrams_results_dict[arr[0]] = [arr[1], arr[2]]
        else:
            trigrams_results_dict[arr[0]] = [arr[1]]
```

```python
counter=0
for line in trigrams_results_dict:
    if counter > 30:
        break
    else:
        print(line, trigrams_results_dict[line])
       
```

```python
import random

def pick_random_next_word(word):
    if word == "<EOS>" or word not in bigrams_results_dict.keys():
        return "<EOS>"
    if word in bigrams_results_dict.keys():
        return random.choice(bigrams_results_dict[word])
    

def generate_random_bigram_sentence():
    result_sentence = []
    current_word = random.choice(list(bigrams_results_dict))
    result_sentence.append(current_word)
    next_word = pick_random_next_word(current_word)
    while next_word != "<EOS>":
        result_sentence.append(next_word)
        current_word = next_word
        next_word = pick_random_next_word(current_word)
    return ' '.join(result_sentence)

    
    

```

```python
for i in range(10):
    print(generate_random_bigram_sentence())
```

```python
def pick_random_next_words(word):
    if word == "<EOS>" or word not in trigrams_results_dict.keys():
        return [" ", "<EOS>"]
    if word in trigrams_results_dict.keys():
        next_words = random.choice(trigrams_results_dict[word])
#         if next_words == "<EOS>":
#             print("PAAAALI SIĘ")
#             return [" ", "<EOS>"]
        if type(next_words) == list:
            return list(next_words)
        else:
            return [" ", next_words]

def generate_random_trigram_sentence():
    result_sentence = []
    current_word = random.choice(list(trigrams_results_dict))
    result_sentence.append(current_word)
    next_words = pick_random_next_words(current_word)
    while next_words[1] != "<EOS>":
        result_sentence.append(next_words[0])
        result_sentence.append(next_words[1])
        current_word = next_words[1]
        next_words = pick_random_next_words(current_word)
    
    result = ' '.join(result_sentence)
    return ' '.join(result.split())


        
```

```python
for i in range(10):
    print(generate_random_trigram_sentence())
```
